```python
# Prepare the feature matrix
x = train_df.iloc[:, 1:17].values
x
```

```python
# Prepare the target variable
y = train_df.iloc[:, 17].values
y
```

Train & Test Split

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

 Logistic Regression

```python
#  Train and evaluate Logistic Regression model
model_lr = LogisticRegression(max_iter = 101)
model_lr.fit(x_train, y_train)
```

```python
model_lr.score(x_test,y_test)
```

```python
y_pred_lr = model_lr.predict(x_test)
```

```python
cr_lr = classification_report(y_test, y_pred_lr)
print(cr_lr)
```

```python
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm,annot=True)
```

K-Nearest Neighbor

```python
# Train and evaluate K-Nearest Neighbors model
scores = {}
slist = []

for i in range(1,10):
    model_knn = KNeighborsClassifier(n_neighbors = i)
    model_knn.fit(x_train, y_train)
    y_pred_knn = model_knn.predict(x_test)
    scores[i] = metrics.accuracy_score(y_test, y_pred_knn)
    slist.append(metrics.accuracy_score(y_test, y_pred_knn))

scores
```

```python
model_knn = KNeighborsClassifier(n_neighbors = 7)
model_knn.fit(x_train, y_train)
```

```python
cr_knn = classification_report(y_test, y_pred_knn)
print(cr_knn)
```

```python
print("Accuracy : ", accuracy_score(y_test, y_pred_knn))
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test, y_pred_knn))
cm_display.plot()
plt.show()
```


Decision Tree

```python
# Train and evaluate Decision Tree model
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
print(f"Accuracy = {metrics.accuracy_score(y_test, y_pred)}")
```

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test, y_pred))
cm_display.plot()
plt.show()
```
